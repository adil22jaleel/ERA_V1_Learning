```python
from __future__ import print_function
import torch
import torch.nn as nn
import torch.nn.functional as F
import torch.optim as optim
from torchvision import datasets, transforms
from torch.optim.lr_scheduler import StepLR
import torchvision
import numpy as np
import matplotlib.pyplot as plt
import os
import warnings
from torchsummary import summary
warnings.filterwarnings("ignore")
from torch.optim.lr_scheduler import StepLR
from main import *
from utils import *
from models import resnet
```

```python
device= get_device()
```

```python
# Loading test and train data & dataloader
cifar10 = dataset_cifar10(batch_size=512)
# Train dataloader
trainloader = cifar10.loader(train_flag=True)
# Test dataloader
testloader = cifar10.loader(train_flag=False)
classes = cifar10.classes
```

```python
cifar10.sample_pictures(train_flag=True)
```

```python
net_18 = resnet.ResNet18().to(device)
summary(net_18, input_size=(3, 32, 32))
```

```python
criterion = nn.CrossEntropyLoss()
optimizer = optim.Adam(net_18.parameters(), lr=1e-7, weight_decay=1e-2)
best_lr=custom_lr_finder(net_18,criterion,optimizer,device,trainloader)
```

```python
lr_scheduler =  torch.optim.lr_scheduler.OneCycleLR( optimizer, max_lr=best_lr, steps_per_epoch=len(trainloader), epochs=20, pct_start=5/20, div_factor=100, three_phase=False, final_div_factor=100, anneal_strategy='linear')
```

```python
history = fit_model(net_18, device, trainloader, testloader,criterion,optimizer,lr_scheduler,NUM_EPOCHS=20)
```

```python
## History contains (training_acc, training_loss, testing_acc, testing_loss)
train_test_plots(history)
```
