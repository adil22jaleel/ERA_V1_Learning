```python
from __future__ import print_function
import torch
import torch.nn as nn
import torch.nn.functional as F
import torch.optim as optim
from torchvision import datasets, transforms
from torch.optim.lr_scheduler import StepLR
import torchvision
import numpy as np
from main.dataset import *
from main.backpropagation import *
from main.custom_resnet import *
from main.utils import print_summary
from main.visualize import *
import matplotlib.pyplot as plt
import os
import warnings
warnings.filterwarnings("ignore")

import torchvision
!pip install torchview
from torchview import draw_graph
import graphviz
```

```python
# CUDA?
cuda = torch.cuda.is_available()
print("CUDA Available?", cuda)
device = torch.device("cuda" if cuda else "cpu")
```

```python


# Loading test and train data & dataloader
cifar10 = dataset_cifar10(batch_size=512)

# Train dataloader
trainloader = cifar10.loader(train_flag=True)

# Test dataloader
testloader = cifar10.loader(train_flag=False)

classes = cifar10.classes
```

```python
cifar10.sample_pictures(train_flag=True)
```

```python
# Loading Custom Model
resnet_model = CustomResNet()

# Sending network to device
model = resnet_model.to(device)
```

```python
print_summary(model, input_size=(3, 32, 32))
```

```python
graphviz.set_jupyter_format('png')
model_graph = draw_graph(CustomResNet(), input_size=(1,3,32,32), expand_nested=False)
model_graph.visual_graph
```

```python
from torch_lr_finder import LRFinder
import torch.nn as nn
import numpy as np


criterion = nn.CrossEntropyLoss()
optimizer = optim.Adam(model.parameters(), lr=1e-7, weight_decay=1e-2)
lr_finder = LRFinder(model, optimizer, criterion, device=device)
lr_finder.range_test(trainloader, end_lr=10, num_iter=100,step_mode="exp")
_, best_lr = lr_finder.plot()
lr_finder.reset()
```

```python
total_epochs = 24
max_lr_epoch = 5
div_factor = 100
final_div_factor = 100
```

```python
print("Max LR is {}".format(best_lr))
```

```python

lr_scheduler =  torch.optim.lr_scheduler.OneCycleLR( optimizer, max_lr=best_lr, steps_per_epoch=len(trainloader), epochs=24, pct_start=5/24, div_factor=100, three_phase=False, final_div_factor=100, anneal_strategy='linear')
```

```python
EPOCHS = 24
for epoch in range(EPOCHS):
    print("EPOCH:", epoch+1)
    train(model, device, lr_scheduler, criterion, trainloader, optimizer, epoch)
    test(model, device, criterion, testloader)
    print("\n")
```
